```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import re

def scrape_marca_english_football():
    # Chromeドライバーのセットアップ
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    try:
        # Marca Englishのフットボールページにアクセス
        url = "https://www.marca.com/en/football.html"
        driver.get(url)

        # ページが完全に読み込まれるまで待機
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "article"))
        )

        # 記事のリストを取得
        articles = driver.find_elements(By.CSS_SELECTOR, "article")

        # 各記事の情報を抽出
        for article in articles:
            try:
                link_element = article.find_element(By.CSS_SELECTOR, "a.ue-c-cover-content__link")
                href = link_element.get_attribute("href")

                # リンク内のh2要素のテキストを取得
                text = link_element.find_element(By.CSS_SELECTOR, "h2.ue-c-cover-content__headline").text

                print(f"HREF: {href}")
                print(f"Text: {text}")
                print("---")

            except Exception as e:
                print(f"Error extracting article info: {e}")

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        # ブラウザを閉じる
        driver.quit()

if __name__ == "__main__":
    scrape_marca_english_football()
```

```python
def scrape_as_english_soccer():
    # Set up Chrome driver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    try:
        # Access AS English soccer page
        url = "https://en.as.com/soccer/"
        driver.get(url)

        # Wait for the page to load completely
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "article"))
        )

        # Get the list of articles
        articles = driver.find_elements(By.CSS_SELECTOR, "article")

        # Extract information from each article
        for article in articles:
            try:
                h2_element = article.find_element(By.CSS_SELECTOR, "h2.s__tl")
                a_element = h2_element.find_element(By.TAG_NAME, "a")
                href = a_element.get_attribute("href")
                text = a_element.text

                print(f"URL: {href}")
                print(f"Title: {text}")
                print("---")
            except Exception as e:
                print(f"Error extracting article info: {e}")

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        # Close the browser
        driver.quit()

if __name__ == "__main__":
    scrape_as_english_soccer()
```

```python
def scrape_sky_english_soccer_top(url):
    # Set up Chrome driver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    try:
        # Access AS English soccer page
        url = "https://www.skysports.com/football"
        driver.get(url)

        # Wait for the page to load completely
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "a.news-top-story__headline-link"))
        )

        # Get the list of articles
        articles1 = driver.find_elements(By.CSS_SELECTOR, "a.news-top-story__headline-link")

        # Get the list of articles
        articles2 = driver.find_elements(By.CSS_SELECTOR, "a.news-list-featured__link")
    
        # Get the list of articles
        articles3 = driver.find_elements(By.CSS_SELECTOR, "a.news-list__headline-link")

        # Get the list of articles
        articles4 = driver.find_elements(By.CSS_SELECTOR, "a.features-block__link")
        
        articles = articles1 + articles2 + articles3 + articles4
             
        # Extract information from each article
        for article in articles:
            try:
                href = article.get_attribute("href")
                text = article.text

                print(f"URL: {href}")
                print(f"Title: {text}")
                print("---")
            except Exception as e:
                print(f"Error extracting article info: {e}")

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        # Close the browser
        driver.quit()

        
if __name__ == "__main__":
    scrape_sky_english_soccer()
```

```python
def scrape_gurdian_english_soccer():
    # Set up Chrome driver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    try:
        # Access AS English soccer page
        url = "https://www.bbc.com/sport/football"
        driver.get(url)

        # Wait for the page to load completely
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "a"))
        )

        # Get the list of articles
        articles = driver.find_elements(By.CSS_SELECTOR, "a")

        # Extract information from each article
        pattern = "https://www.bbc.com/sport/football/articles/"
                
        for article in articles:
                try:
                    href = article.get_attribute("href")
                    text = article.text
                    if re.match(pattern, href):
                        if 'comments' not in href:
                            print(f"URL: {href}")
                            print(f"Title: {text}")
                            print("---")
                except Exception as e:
                    print(f"Error extracting article info: {e}")

    except Exception as e:
        print(f"An error occurred: {e}")

    finally:
        # Close the browser
        driver.quit()

if __name__ == "__main__":
    scrape_gurdian_english_soccer()
```
